```python
choice = input("Choose operator: ") # selecting the operator
num1 = float(input("First number: "))
num2 = float(input("Second number: "))
if choice == '+':
    print(num1 + num2)
    
elif choice == '-':
    print(num1 - num2)
    
elif choice == '*':
    print(num1 * num2)
    
elif choice == '/':
    print(num1 / num2)
    
elif choice == '//':
    print(num1 // num2)
    
elif choice == '%':
    print(num1 % num2)
    
elif choice == '**':
    print(num1 ** num2)
    
else:
    print("Invalid operation!")
```
