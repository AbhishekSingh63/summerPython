```python
#Q1
for i in range(1,6):
    for j in range(1):
        print("*")
```

```python
#Q2
for i in range(4):
    for j in range(5):
        print("*",end="")
    print()
```

```python
#Q3
for i in range(1,6):
    for j in range(1,i+1):
        print("*",end="")
    print()
```

```python
#Q4
n=int(input("number"))
for i in range(1,n+1):
    for j in range(n-i+1):
        print("*",end="")
    print()
```

```python
#Q5
for i in range(6):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
#Q6
n=int(input("number"))
for i in range(n):
    for j in range(1,n-i+1):
        print(j,end="")
    print()
```
