```python
gender=input("Enter your Gender: ")
group=input("Enter your age group: ")
total=int(input("Total purchase amount: "))

#offer on clothes for men
if gender=="male":
    if group=="child" and total>=1000:
        print("buy 1 get 1 offer on any purchase")
    elif group=="teen" and total>=2000:
        print("50% Discount recieved")
    elif group=="adult" and total>=2500:
        print("30% Discount recieved")
    else:
        print("no valid offers")
#offer on clothes for women
elif gender=="female":
    if group=="child" and total>=1000:
        print("buy 1 get 1 offer on any purchase")
    elif group=="teen" and total>=2000:
        print("60% Discount recieved")
    elif group=="adult" and total>=3000:
        print("40% Discount recieved")
    else:
        print("no valid offers")
else:
    print("invalid gender")
    
print("Happy Shopping")
```
