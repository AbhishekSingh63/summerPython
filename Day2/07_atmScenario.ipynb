```python
lang = input("Enter preferred language to proceed: ")
acc = input("Choose account: ")

if lang == "english":
    if acc == "savings":
        amount = int(input("Enter amount: "))
        pin = int(input("Enter PIN: "))

        if 100000 <= pin <= 999999:
            print("Withdrawal complete")
        else:
            print("Invalid PIN")

    elif acc == "current":
        print("Current account withdrawal not working!")

elif lang == "hindi":
    if acc == "bajat khata":
        amount = int(input("Paise daale: "))
        pin = int(input("Apna bank PIN daale: "))

        if 100000 <= pin <= 999999:
            print("Apke paise nikal gaye")
        else:
            print("Galat PIN")

    elif acc == "current":
        print("Yeh service available nahi hai!")

else:
    print("Language not selected")
```
